```python
# 1. Obtenga la distribución de cada una de las variables

datos.describe()
```

```python
# graficar la distribución de cada una de las variables

fig, axs = plt.subplots(1, 4 , figsize=(25, 5))
datos.hist(column='tv', ax=axs[0])
datos.hist(column='radio', ax=axs[1])
datos.hist(column='periodico', ax=axs[2])
datos.hist(column='ventas', ax=axs[3])

axs[0].set_title('Distribución de tv')
axs[1].set_title('Distribución de radio')
axs[2].set_title('Distribución de periodico')
axs[3].set_title('Distribución de ventas')
```

```python
# 2. Separe sus datos en 70/30

X = datos[['tv', 'radio', 'periodico']]
y = datos['ventas']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
# 3. Obtenga su modelo de Regresión lineal

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
# 4. Entregue la recta matemática de regresión con las estimaciones obtenidas

print('Intercepto: ', round(modelo.intercept_, 3)) 
print('Coeficientes: ', np.round(modelo.coef_, 3)) 

# ecuación de la recta
print('y = ', 
      round(modelo.intercept_, 3), ' + ', # intercepto
      round(modelo.coef_[0], 3), 'x1 + ', # coeficiente de x1
      round(modelo.coef_[1], 3), 'x2 + ', # coeficiente de x2
      round(modelo.coef_[2], 3), 'x3')    # coeficiente de x3

```

```python
# 5. Realice un diagnóstico de los residuos del modelo

# predicciones
prediccion = modelo.predict(X_test)

# residuos
residuos = y_test - prediccion
residuos.head()
```

```python
# gráfico de residuos
plt.scatter(prediccion, residuos)
plt.axhline(y=0, color='red', label='residuo = 0')
plt.xlabel('y_pred')
plt.ylabel('residuo')
plt.legend()
plt.show()

# histograma de residuos
plt.hist(residuos)
plt.xlabel('residuo')
plt.ylabel('frecuencia')
plt.show()
```

```python
# 6. Realice el test de normalidad de los residuos

# test de normalidad
_, p = shapiro(residuos)

print(f'Shapiro-Wilk Test: {p:.5f}')

if p > 0.05:
    print('Se rechaza H0, por lo que se puede asumir que los residuales del modelo distribuyen de forma normal.')

else:
    print('No se rechaza H0, por lo que los residuales del modelo no distribuyen de forma normal.')

```

```python
# 7. Obtenga las predicciones del modelo

# predicciones
prediccion = modelo.predict(X_test)

# graficar las predicciones con axs

fig, axs = plt.subplots(1, 3, figsize=(25, 5))

axs[0].scatter(X_test['tv'], y_test, color='blue', label='real')
axs[0].scatter(X_test['tv'], prediccion, color='red', label='predicción')
axs[0].set_xlabel('tv')
axs[0].set_ylabel('ventas')
axs[0].legend()

axs[1].scatter(X_test['radio'], y_test, color='blue', label='real')
axs[1].scatter(X_test['radio'], prediccion, color='red', label='predicción')
axs[1].set_xlabel('radio')
axs[1].set_ylabel('ventas')
axs[1].legend()

axs[2].scatter(X_test['periodico'], y_test, color='blue', label='real')
axs[2].scatter(X_test['periodico'], prediccion, color='red', label='predicción')
axs[2].set_xlabel('periodico')
axs[2].set_ylabel('ventas')
axs[2].legend()

plt.show()
```

```python
# 8. Entregue el error del test

# error del test
error = mean_squared_error(y_test, prediccion)
print('Error del test: ', error)

```
